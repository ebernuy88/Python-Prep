```python
#1.ENTRADA DE DATOS
#  Se refiere al proceso de obtener datos de los usuarios (entrada)

#A.Función input()
#  Muestra un mensaje al usuario (opcional) y espera a que el usuario escriba algo
#  El valor ingresado por el usuario se devuelve como una cadena de texto.


nombre = input("Por favor, ingresa tu nombre: ")

#B.CONVERSION A OTRO TIPO DE DATOS
#  Puedes convertir la entrada utilizando las funciones int() o float().

edad = int(input("Por favor, ingresa tu edad: "))
altura = float(input("Por favor, ingresa tu altura en metros: "))
```

```python
#2. SALIDA DE DATOS
#   Mostrar resultados o información al usuario 

#A.Función print()
#  Se utiliza para mostrar información al usuario en la consola

print("Hola, mundo!")

#B.Formato de cadena
#  Puedes formatear las cadenas de salida utilizando diferentes métodos

#1.Concatenación de cadenas
nombre = "Juan"
edad = 30
saludo = "Hola, mi nombre es " + nombre + " y tengo " + str(edad) + " años."
print(saludo)

#2.Formato de cadena antiguo (%)
#utilizamos %s y %d como marcadores de posición Luego, usamos el operador % para reemplazar esos marcadores de posición 
#con los valores de nombre y edad.

nombre = "María"
edad = 25
saludo = "Hola, mi nombre es %s y tengo %d años." % (nombre, edad)
print(saludo)


#3.Método format()

nombre = "Pedro"
edad = 40
saludo = "Hola, mi nombre es {} y tengo {} años.".format(nombre, edad)
print(saludo)

#utilizamos llaves {} como marcadores de posición dentro de la cadena, y luego llamamos al método format() 
#para reemplazarlos con los valores de nombre y edad.

#4.Literales de cadena
#podemos incluir variables directamente en la cadena, precedidas por el prefijo f
nombre = "Ana"
edad = 35
saludo = f"Hola, mi nombre es {nombre} y tengo {edad} años."
print(saludo)

#5.Otra salida

nombre = "Juan"
edad = 25
print("Mi nombre es", nombre, "y tengo", edad, "años.")
```
